```
import matplotlib.pyplot as plt
import pandas as pd
dt=pd.read_csv(url, error_bad_lines=False)

#Plots in matplotlib reside within a figure

fig=plt.figure()


ax = fig.add_subplot(1,1,1)

#Variable
ax.hist(dt['Country'],bins = 10)
#Labels and Tit
plt.title('Data Statistik lalu lintas bandara yang terdampak Covid-19')
plt.xlabel('Negara')
plt.ylabel('Jumlah')
plt.show()
```

pada data yang telah di analisis menunjukkan diagram histogram yang tidak normal, data tersebut adalah jumlah seluruh negara yang dihasilkan dengan total data statistik lalulintas bandara yang terdampak covid-19

```
fig=plt.figure()

ax = fig.add_subplot(1,1,1)

ax.scatter(dt['Country'],dt['City'])

plt.title('Data lalu lintas bandara yang terdampak Covid-19')
plt.xlabel('Negara')
plt.ylabel('Kota')
plt.show()
```

hasil dari scratter plot yang telah dianalisis menunjukan nilai x semakin besar dan semakin besar pula nilai y. Dapat disimpulkan variabel yang diteliti memiliki hubungan linear atau positif dan memiliki hubungan yang sangat kuat.

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dt = pd.read_csv(url, error_bad_lines=False)

dt = dt.pivot_table(index='AggregationMethod', columns='City', values='PercentOfBaseline')

print (dt)

sns.heatmap(dt, cmap='coolwarm')

plt.show()
```
